9)Company_size: The size of the companies, classified as "L" (Large), "M" (Medium), and "S" (Small).

```python
import re

def resumir_titulo(titulo):
    # Definir patrón regex para extraer la parte relevante del título
    patron = re.compile(r'\b(Data Engineer|Data Science|Data Scientist|Machine Learning|Data Analyst|Data Architect)\b', flags=re.IGNORECASE)

    # Buscar el patrón en el título
    coincidencias = patron.findall(titulo)

    try:
      output=coincidencias[0]
    except:
      output=titulo
    return output
```

```python
salario['job_title']=salario.job_title.apply(resumir_titulo)
```

```python
salario.job_title.value_counts()
```

```python
salario['job_title'] = salario['job_title'].replace({'Data Analytics': 'Data Analyst',
                                                               'Data Operations Analyst':'Data Analyst',
                                                               'Data Analytics Consultant':'Data Analyst',
                                                               'Data Analytics Lead':'Data Analyst',
                                                               'ML Engineer':'Machine Learning',
                                                               'Data Quality Analyst':'Data Analyst',
                                                               'Data Analytics Manager':'Data Analyst',
                                                               'Data Infrastructure Engineer':'Data Engineer'})
```

```python
df=salario
df.head()
```

```python
X=df.drop(['salary_in_usd'],axis=1)
X
```

```python
y=df.salary_in_usd
y
```

**Utilizando todas las variables el r2 no es bueno, por lo que se puede observar que se esta overfitiando.**

```python
# Divido el dataset en test y entrenamiento
X_train, X_test, Y_train, Y_test = train_test_split(X, y, random_state=42)
```

```python
df.dtypes
```

```python
lista_numeric_features = df.select_dtypes(include=['int64']).drop(['salary_in_usd'], axis=1).columns
```

```python
lista_numeric_features
```

```python
lista_categorical_features = df.select_dtypes(include=['object']).columns
```

```python
lista_categorical_features
```

```python
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])
```

```python
categorical_transformer = Pipeline(steps=[
      ('imputer', SimpleImputer(strategy='most_frequent')),
      ('onehot', OneHotEncoder(handle_unknown='ignore'))])
```

```python
Preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, lista_numeric_features),
        ('cat', categorical_transformer, lista_categorical_features)])
```

```python
Preprocessor
```

```python
Preprocessor.fit_transform(X_train)
```

```python
pipe = Pipeline(steps=[('preprocessor', Preprocessor),
                      ('modelo', LR())])
```

```python
pipe
```

Defino la grilla con las diferentes modelos e hiperparámetros:

```python
params_grid = [

    {
       "modelo":[LR()]
    },


    {
     "modelo" :[DecisionTreeRegressor()],
     "modelo__max_depth": [10, 20, 30]
    },


    {
     "modelo" :[RandomForestRegressor()],
     "modelo__n_estimators": [50, 100, 200]
    },


    {
     "modelo" :[XGBRegressor()],
     "modelo__n_estimators": [50, 100, 200],
     "modelo__learning_rate": [0.01, 0.1, 0.3],

    },

    {
     "modelo" :[CatBoostRegressor()],
     "modelo__n_estimators": [5, 10, 20],
     "modelo__learning_rate": [0.01, 0.1, 0.3]
    },

    {
     "modelo" :[LGBMRegressor()],
     "modelo__n_estimators": [50, 100, 200],
     "modelo__learning_rate": [0.01, 0.1, 0.3]
    },


]
```

Instancio utilizando la grilla y cross validation

```python
pipe_search_cv=GridSearchCV(pipe, params_grid, cv=5, verbose=0, n_jobs=4)
```

```python
pipe_search_cv.fit(X_train, Y_train)
#el pipeline ganador
pipe_search_cv.best_estimator_
```

```python
pipe_search_cv.best_params_
```

```python
pipe_final=pipe_search_cv.best_estimator_
```

```python
pipe_final.fit(X_train,Y_train)
```

```python
y_pred_test=pipe_search_cv.predict(X_test)
```

```python
#Calculo las métricas para validar el modelo

rmse = np.sqrt(mean_squared_error(Y_test, y_pred_test))
r2 = r2_score(Y_test, y_pred_test)
print(f"RMSE: {rmse}")
print(f"R2 Score: {r2}")
```

##Conclusión


Después de probar varios modelos de regresión utilizando técnicas de GridSearchCV, se encontró que el mejor modelo para predecir los salarios en dólares fue el CatBoostRegressor, con un  R2 de 0.47. Sin embargo, el rendimiento del modelo sigue sin ser lo esperable. Esto puede atribuirse a la poca cantidad de datos de entrenamiento y la presencia de categorías con datos limitados, lo que puede dificultar la generalización del modelo. Aunque el CatBoostRegressor pudo capturar algunas relaciones en los datos, se necesitarían más datos de entrenamiento y una mayor cantidad de datos de las diferentes categorías para mejorar aún más el rendimiento del modelo y hacer predicciones más precisas de los salarios en dólares.
